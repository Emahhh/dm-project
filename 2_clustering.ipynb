```python
import pandas as pd
import seaborn as sns

cyclists_df = pd.read_pickle('cyclists_transformed.pkl')
cyclists_df.info()
```

```python
races_df = pd.read_pickle('races_transformed.pkl')
races_df.info()
```

```python
sns.pairplot(cyclists_df)
```

```python
sns.pairplot(races_df)
```

```python
merged_df = pd.merge(races_df, cyclists_df, left_on='cyclist', right_on='_url')
merged_df.head()
```

```python
merged_df.info()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming merged_df is already defined
# Select relevant attributes
relevant_attributes = ['points', 'length', 'climb_total', 'profile', 'cyclist_age', 'delta', 'weight', 'height', 'BMI']

# Drop rows with missing values in relevant attributes
filtered_df = merged_df[relevant_attributes].dropna()

# Sample a few points
sampled_df = filtered_df.sample(n=500, random_state=42)

# Print a pairplot
sns.pairplot(sampled_df)
plt.show()
```
